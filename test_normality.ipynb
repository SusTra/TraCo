## Normality tests
https://machinelearningmastery.com/a-gentle-introduction-to-normality-tests-in-python/

```python
def shapiro_wilk(data):
    # normality test
    stat, p = shapiro(data)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print('Sample looks Gaussian (fail to reject H0)')
    else:
        print('Sample does not look Gaussian (reject H0)')

def dagostino_pearson(data):
    # normality test
    stat, p = normaltest(data)
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print('Sample looks Gaussian (fail to reject H0)')
    else:
        print('Sample does not look Gaussian (reject H0)')

def anderson_darling(data):
    result = anderson(data)
    print('Statistic: %.3f' % result.statistic)
    p = 0
    for i in range(len(result.critical_values)):
        sl, cv = result.significance_level[i], result.critical_values[i]
        if result.statistic < result.critical_values[i]:
            print('%.3f: %.3f, data looks normal (fail to reject H0)' % (sl, cv))
        else:
            print('%.3f: %.3f, data does not look normal (reject H0)' % (sl, cv))
    
```

## Data import

```python
df1 = pd.read_csv("data\\telraam.csv")
df2 =pd.read_csv("data\\counters.csv")
df = pd.concat([df1, df2], ignore_index=True)
df.head()
```

```python
for counter_id in df.counter_id.unique():
    data = df.loc[df.counter_id == counter_id,'count'].values
    
    print("*****")
    print("*****")
    print(counter_id)
    
    plt.hist(data)
    plt.title(counter_id)
    plt.show()
    
    qqplot(data, line='s')
    plt.title(counter_id)
    plt.show()
        
    print("*****")
    print("Shapiro-Wilk Test")
    shapiro_wilk(data)
    print("*****")
    print("D’Agostino’s K^2 Test")
    dagostino_pearson(data)
    print("*****")
    print("Anderson-Darling Test")
    anderson_darling(data)
    
    
    
```
